```python
import pandas as pd
```

```python
YEAR = 2016
```

```python
df = pd.read_csv(f'downloads/es_{YEAR}.zip', delimiter=';', encoding='latin1', compression='zip')
df['year'] = YEAR
df = df.rename(columns={'COMARCA': 'MUNICIPIO'})
df.head()
```

```python
cols = 'BENEFICIARIO	PROVINCIA	MUNICIPIO	MEDIDA'.split()
for c in cols:
    df[c] = df[c].str.replace(r"^'(.*)'$", '\\1')
df.head()
```

```python
df_muni = pd.read_csv(f'downloads/es_{YEAR}_muni.zip', delimiter=';', encoding='latin1', compression='zip')
df_muni['year'] = YEAR
df_muni = df_muni.rename(columns={'COMARCA': 'MUNICIPIO'})
df_muni.head()
```

```python
cols = 'BENEFICIARIO	PROVINCIA	MUNICIPIO	MEDIDA'.split()
for c in cols:
    df[c] = df[c].str.replace(r"^'(.*)'$", '\\1')
df.head()
```

```python
len(df), len(df_muni)
```

```python
df = pd.concat([df, df_muni])
```

```python
df['amount'] = pd.to_numeric(df['IMPORTE_EUROS'].str.replace(',', '.'))
```

```python
df['recipient_postcode'] = df['MUNICIPIO'].str.split(' - ').apply(lambda x: x[0])
df.head()
```

```python
df['recipient_location'] = df.apply(lambda x: '%s, %s' % (x['MUNICIPIO'].split(' - ')[1], x['PROVINCIA']), 1)
df.head()
```

```python
from slugify import slugify

df = df.rename(columns={
    'BENEFICIARIO': 'recipient_name',
    'MEDIDA': 'scheme',
})
df = df.drop(['PROVINCIA', 'MUNICIPIO', 'IMPORTE_EUROS'], 1)
df['country'] = 'ES'
df['currency'] = 'EUR'
df['recipient_id'] = df.apply(lambda x: 'ES-%s-%s' % (x['recipient_postcode'], slugify(x['recipient_name'])), 1)
df.head()
```

```python
df.to_csv('es_2016.csv.gz', index=False, encoding='utf-8', compression='gzip')
```
